```python
n = 10
```

```python
for i in range(1, n+1):
    G = nx.generators.random_graphs.powerlaw_cluster_graph(150, 4, 0.05)
    d = GC.betweenness_centrality(G)

    nx.write_edgelist(G, "./test_data/syn_data/10000/orign/test_edge10000_{}".format(i))
    with open("./test_data/syn_data/10000/orign/test_betweenness10000_{}".format(i).format(i), 'wb') as f:
        pickle.dump(d, f)

    myG = fc.myGraph()
    myG.nodeProcess(G=G)
    myG.scoreProcess(G_dict=d)
    myG.degree_Coeffient()
    
    with open('./test_data/syn_data/10000/test_G{}'.format(i), 'wb') as f:
        pickle.dump(myG, f)

    del(G)
    del(d)
    del(myG)
```
